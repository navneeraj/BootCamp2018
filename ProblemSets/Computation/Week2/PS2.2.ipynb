```python
# Problem 2
x = np.linspace(0,1,100)
def bern(n,v):
    coeff = st.special.binom(n,v)*(x**v)*((1-x)**(n-v))
    return coeff

N = 4
fig, ax = plt.subplots(N,N)
for nn in range(N):
    for vv in range(nn+1):
        ax[nn,vv].plot(x, bern(nn,vv))
plt.show()
plt.tight_layout()
```

```python
# Problem 3
bball = np.load("MLB.npy")
fig, ax = plt.subplots(1,3)
ax[0].scatter(bball[:,0],bball[:,1])
reg1 = st.stats.linregress(bball[:,0],bball[:,1])
ax[0].plot(bball[:,0],bball[:,0]*reg1[0]+reg1[1] ,'r--')
ax[0].set_xlabel("height")
ax[0].set_ylabel("weight")
ax[1].scatter(bball[:,1],bball[:,2])
reg2 = st.stats.linregress(bball[:,1],bball[:,2])
ax[1].plot(bball[:,1],bball[:,1]*reg2[0]+reg2[1] ,'r--')
ax[1].set_xlabel("weight")
ax[1].set_ylabel("age")
ax[2].scatter(bball[:,2],bball[:,0])
reg3 = st.stats.linregress(bball[:,2],bball[:,0])
ax[2].plot(bball[:,2],bball[:,2]*reg3[0]+reg3[1] ,'r--')
ax[2].set_xlabel("age")
ax[2].set_ylabel("height")
plt.tight_layout()
```
