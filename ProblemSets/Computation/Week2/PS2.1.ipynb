```python
import numpy as np
from matplotlib import pyplot as plt

def prob1_1(n):
    return np.var(np.mean(np.random.normal(size = (n,n)),axis =1))


nvec = np.array(range(100,1001,100))
def prob1_2():
    x=[]
    for i in range(len(nvec)):
        x =  np.append(x, prob1_1(nvec[i]))
    z =  plt.plot(nvec,x)
prob1_2()
```

```python
def prob_2():
    x  = np.linspace(-2*np.pi,2*np.pi, 1000)
    plt.plot(x, np.sin(x))
    plt.plot(x, np.cos(x))
    plt.plot(x, np.arctan(x))
prob_2()
```

```python
def prob_3():
    x1  = np.linspace(-2.0,0.9, 100) 
    x2  = np.linspace(1.1,6.0, 100) 
    y1 = (1/(x1-1))
    y2 = (1/(x2-1))
    plot1 = plt.plot(x1,y1, 'm--', linewidth =4)
    plot2 = plt.plot(x2,y2, 'm--', linewidth =4)
    plt.xlim(-2, 6)
    plt.ylim(-6, 6)

prob_3()
```

```python
def prob_4():
    x  = np.linspace(0,2*np.pi, 1000)
    plt.suptitle("Sinusoidal")
    plt.axis([0, 2*np.pi,-2,2])
    y = plt.subplot(221)
    y.plot(x,np.sin(x),'g')
    y.set_title("Sin(x)")
    z = plt.subplot(222)
    z.plot(x,np.sin(2*x),'r--')
    z.set_title("Sin(2x)")
    a = plt.subplot(223)
    a.plot(x,2*np.sin(x),'b--')
    a.set_title("2Sin(x)")
    b = plt.subplot(224)
    b.plot(x,2*np.sin(2*x),'m.')
    b.set_title("2Sin(2x)")
prob_4()
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = x.copy()
X, Y = np.meshgrid(x, y)
Z = (np.sin(X) * np.sin(Y))/(X*Y) 
# Plot the heat map of f over the 2-D domain.
plt.subplot(121)
plt.pcolormesh(X, Y, Z, cmap="viridis")
plt.colorbar()
plt.xlim(-2*np.pi, 2*np.pi)
plt.ylim(-2*np.pi, 2*np.pi)
plt.subplot(122)
plt.contourf(X, Y, Z, 10, cmap="coolwarm")
plt.colorbar()
plt.xlim(-2*np.pi, 2*np.pi)
plt.ylim(-2*np.pi, 2*np.pi)
plt.tight_layout()
```
