```python
import numpy as np
import scipy as scp
import matplotlib.pyplot as plt
```

# Problem 8.1

```python
x = np.linspace(0,50,10000)

y1 = (2*x+4)/3
y2 = (x-1)/6
y3 = 6-x
plt.plot(x,y1 ,'r')
plt.plot(x, y2,'g')
plt.plot(x, y3, 'b')
plt.plot(x, (5*x/4),'m--')
plt.plot(x, (5*x-2)/4,'m--')
y4 = np.maximum(y1,y2)
y5 = np.maximum(y3,y4)
plt.fill_between(x,y5,y4,color = 'grey',alpha = 0.3)
plt.xlim([0,6])
plt.ylim([0,6])

plt.plot(14/5, 16/5,'.', markersize= 15)
```

We know from the fundamental theorem of linear optimization that the optimal will lie on a vertex.
In the above question,we can see that there is only one vertex for the feasible set and the hyperplanes given in the problem. Therefore the solution will be intersection of the blue line ($x+y =6$) and red line ($2x-3y =-4$). This is given by the point $\mathbf{x = \frac{14}{5} ,y = \frac {16}{5}}$.

# Problem 8.2(i)

```python
x = np.linspace(0,50,10000)

y1 = (15-x)/3
y2 = (18-2*x)/3
y3 = x-4
plt.plot(x,y1 ,'r')
plt.plot(x, y2,'g')
plt.plot(x, y3, 'b')
plt.plot(x,10 -3*x,'m--')
plt.plot(x,20 -3*x,'m--')
# plt.plot(x, (5*x-2)/4,'m--')
y4 = np.minimum(y1,y2)
y5 = np.minimum(y3,y4)
plt.fill_between(x,y4,y5,color = 'grey',alpha = 0.3)
plt.xlim([0,15])
plt.ylim([0,6])
plt.plot(6, 2,'.', markersize= 15)
```

We know from the fundamental theorem of linear optimization that the optimal will lie on a vertex. In the above question,we can see that there are two vertices for the feasible set and the hyperplanes given in the problem. In the optimization $x_1$ has three times weight than $x_2$. Hence solution will lie where $x_1$> $x_2$, which is intersection of the blue line ( $ x_1 -  x_2 =4 $ ) and red line ( $ 2x_1 + 3x_2 = 18$ ). This is given by the point  $ \mathbf{x_1=6, x_2=2}$ .

# Problem 8.2(ii)

```python
x = np.linspace(0,50,10000)
y1 = 11+x
y2 = 27-x
y3 = (90-2*x)/5
plt.plot(x,y1 ,'r')
plt.plot(x, y2,'g')
plt.plot(x, y3, 'b')
plt.plot(x, (100-4*x)/6,'m--')
plt.plot(x, (150-4*x)/6,'m--')
plt.fill_between( x,0, y1, color='grey', where = y3 >= y1, interpolate = True, alpha = 0.3)
plt.fill_between(x,0, y2, color='grey', where = y2 <= y3, interpolate = True, alpha = 0.3)
plt.fill_between(x,0,y3, color='grey', where = (y3 <= y2) & (y3 <= y1), interpolate = True, alpha = 0.3)
plt.xlim([0,50])
plt.ylim([0,25])

sys13 = np.array([-1, 1,2,5]).reshape(2,2)
b13 = np.array([11,90])
xcv =np.linalg.solve(sys13,b13)
print(xcv, 4*xcv[0]+ 6*xcv[1])

sys23 = np.array([1, 1,2,5]).reshape(2,2)
b23 = np.array([27,90])
xcv1 =np.linalg.solve(sys23,b23)
print(xcv1, 4*xcv1[0]+ 6*xcv1[1])

plt.plot(xcv1[0], xcv1[1],'.', markersize= 15)
```

# Problem 8.7(i)- Feasible set exists but origin is not feasible

```python
x = np.linspace(0,50,10000)
y1 = (-4*x+8)/2
y2 = (6+2*x)/3
y3 = np.linspace(0,50,10000)
plt.plot(x,y1 ,'r')
plt.plot(x, y2,'g')
plt.plot(3*np.ones_like(x), y3, 'b')
plt.plot(x, (6-x)/2,'m--')
plt.plot(x, (11-x)/2,'m--')
# plt.plot(x, (150-4*x)/6,'m--')
plt.fill_between(x, y1, y2, color='grey', where = (y2 >= y1) & (y3<=3), interpolate = True, alpha = 0.3)
plt.xlim([0,5])
plt.ylim([0,5])
plt.plot(3, 4,'.', markersize= 15)
```
